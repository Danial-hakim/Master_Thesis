```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
seen_data = {
    'YOLOv11': {'precision': 0.88, 'recall': 0.69, 'mAP@50': 0.75, 'mAP@50-95': 0.48},
    'YOLOv5':   {'precision': 0.81, 'recall': 0.53, 'mAP@50': 0.60, 'mAP@50-95': 0.31},
    'YOLOv8':   {'precision': 0.85, 'recall': 0.62, 'mAP@50': 0.69, 'mAP@50-95': 0.39}
}

unseen_data = {
    'YOLOv11': {'precision': 0.90, 'recall': 0.05, 'mAP@50': 0.10, 'mAP@50-95': 0.05},
    'YOLOv5':   {'precision': 0.44, 'recall': 0.14, 'mAP@50': 0.18, 'mAP@50-95': 0.10},
    'YOLOv8':   {'precision': 0.80, 'recall': 0.16, 'mAP@50': 0.28, 'mAP@50-95': 0.13}
}

metrics = ['precision', 'recall', 'mAP@50', 'mAP@50-95']
```

```python
colors = ['blue', 'red', 'green']  
def plot_single_dataset(data, title):
    models = list(data.keys())
    x = np.arange(len(metrics))
    width = 0.2
    
    fig, ax = plt.subplots(figsize=(8, 5))
    for i, model in enumerate(models):
        vals = [data[model][m] for m in metrics]
        ax.bar(x + (i - 1) * width, vals, width, label=model, color=colors[i])
    
    ax.set_ylabel('Score', fontsize=16)
    ax.set_title(title, fontsize=16)
    ax.set_xticks(x)
    ax.set_xticklabels(metrics, fontsize=15)
    ax.set_yticks(np.linspace(0, 1, 6))
    ax.set_yticklabels([f"{y:.1f}" for y in np.linspace(0, 1, 6)], fontsize=14)
    ax.legend(fontsize=15)
    plt.tight_layout()
    plt.show()
```

```python
plot_single_dataset(seen_data, 'Seen Data Performance for YOLO Models')
```

```python
plot_single_dataset(unseen_data, 'Unseen Data Performance for YOLO Models')
```

```python
seen_inference_speed = {
    'YOLOv11': 1.9,
    'YOLOv5': 1.7,
    'YOLOv8': 1.7
}

unseen_inference_speed = {
    'YOLOv11': 2.9,
    'YOLOv5': 3.6,
    'YOLOv8': 2.5
}
```

```python
models = list(seen_inference_speed.keys())
x = np.arange(len(models))
width = 0.3

fig, ax = plt.subplots(figsize=(8, 5))

# Bars
bars_seen = ax.bar(x - width/2,
                   list(seen_inference_speed.values()),
                   width,
                   label='Seen Data',
                   color=colors,
                   alpha=0.8)

bars_unseen = ax.bar(x + width/2,
                     list(unseen_inference_speed.values()),
                     width,
                     label='Unseen Data',
                     color=colors,
                     alpha=0.4)

# Add labels above bars
for bars in [bars_seen, bars_unseen]:
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 0.05,
                f"{bar.get_height():.2f} ms",
                ha='center', va='bottom')

# Labels and title
ax.set_ylabel("Inference Speed (ms)", fontsize=16)
ax.set_title("Seen vs Unseen Inference Speed by YOLO Model", fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=15)
ax.legend(fontsize=15)
plt.tight_layout()
plt.show()
```

```python
result_100 = "runs/detect/train_yolo11_n21/results.csv"
result_200 = "runs/detect/train_yolo11_n24/results.csv"

df_100 = pd.read_csv(result_100)
df_200 = pd.read_csv(result_200)
```

```python
epoch_col = next(c for c in df_100.columns if c.strip().lower().startswith("epoch"))
prec_col  = next(c for c in df_100.columns if "metrics/precision" in c.lower() or "precision" in c.lower())
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_100[epoch_col], df_100[prec_col], linestyle='--', label='Precision (100 epochs)')
plt.plot(df_200[epoch_col], df_200[prec_col], linestyle='-',  label='Precision (200 epochs)')
plt.xlabel("Epochs")
plt.ylabel("Precision")
plt.title("Precision vs Epochs: 100 vs 200")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def plot_metric(df_100, df_200, metric_keyword, label_100="100 epochs", label_200="200 epochs"):
    # Find epoch column
    epoch_col = next(c for c in df_100.columns if c.strip().lower().startswith("epoch"))
    # Find metric column
    metric_col = next(
        c for c in df_100.columns 
        if metric_keyword.lower() in c.lower()
    )

    plt.figure(figsize=(8,5))
    plt.plot(df_100[epoch_col], df_100[metric_col], linestyle='--', label=f'{metric_keyword.title()} ({label_100})')
    plt.plot(df_200[epoch_col], df_200[metric_col], linestyle='-', label=f'{metric_keyword.title()} ({label_200})')
    plt.xlabel("Epochs")
    plt.ylabel(metric_keyword.title())
    plt.title(f"{metric_keyword.title()} vs Epochs: {label_100} vs {label_200}")
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_metric(df_100, df_200, "precision")
```

```python
plot_metric(df_100, df_200, "recall")
```

```python
plot_metric(df_100, df_200, "map50")
```

```python
plot_metric(df_100, df_200, "map50-95")
```
